```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
```

```python
b = a**4
```

```python
x = np.arange(0, 10)
```

```python
y = 2*x
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(a, b)
axes[1].plot(x, y)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6,3), dpi=200)
axes[0][0].plot(x, y)
axes[0][0].set_title('Some Title')
axes[1][0].plot(x, y)
axes[1][1].plot(x, y)
axes[0][1].plot(a, b)
#plt.tight_layout()
fig.subplots_adjust(wspace=0.9, hspace=0.9)
fig.suptitle('Super Level Title', fontsize=14)
plt.show()
fig.savefig('new_subplots.png', bbox_inches='tight')
```
