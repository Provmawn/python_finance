```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
m = np.linspace(0, 10, 11)
```

```python
c = 3e8
```

```python
E = m*(c**2)
```

```python
E
```

```python
plt.plot(m, E, color='red', lw=5)
plt.title('E=mc^2')
plt.xlabel('M in g')
plt.ylabel('E in J')
plt.xlim(0, 10)
plt.show()
```

```python
plt.plot(m, E, color='red', lw=5)
plt.title('E=mc^2')
plt.xlabel('M in g')
plt.ylabel('E in J')
plt.xlim(0, 10)
plt.yscale('log')
plt.grid(which='both', axes='y')
```
