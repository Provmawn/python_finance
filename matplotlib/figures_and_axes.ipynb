```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b)
plt.show()
```

```python
fig1 = plt.figure()

axes1 = fig1.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

axes2 = fig1.add_axes([.2, .5, .25, .25])
axes2.plot(a, b)
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 6), dpi=500)
axes = fig.add_axes([0,0, 1, 1])
axes.plot(a, b)

fig.savefig('new_figure.png', bbox_inches='tight')
```
