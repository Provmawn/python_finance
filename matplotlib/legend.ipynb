```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, label='X vs X')
axes.plot(x, x**2, label='X vs X^2')
axes.legend(loc=(1.1, 0.2))
```
