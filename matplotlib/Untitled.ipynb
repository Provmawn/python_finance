```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10000, 10000)
```

```python
y = (x*4+x*2+1)/(x*2-1)
```

```python
plt.plot(x, y)
plt.title("Ryan's Function")
plt.xlim(-100, 100)
plt.savefig('plots/cool_func.png')
```
