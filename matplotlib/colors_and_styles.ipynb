```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, lw=5, ls='--',color='#650dbd', label='x vs x')
axes.plot(x, x+1, color='purple', label='x vs x+1')
axes.legend(loc=(1.1,.2))
plt.show()
```
