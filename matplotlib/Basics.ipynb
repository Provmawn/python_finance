```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(x.min(), x.max())
plt.ylim(x.min(), y.max())
plt.title('String Title')
plt.savefig('plots/myfirstplot.png')
```
