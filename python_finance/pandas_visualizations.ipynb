```python
import numpy as np
import pandas as pd
import datetime as datetime
```

```python
df = pd.read_csv('datasets/FB.csv', parse_dates=[0], index_col=0)
```

```python
df.info()
```

```python
df['Close/Last'] = df['Close/Last'].apply(lambda s: s.replace('$', ''))
```

```python
df.info()
```

```python
df['Close/Last'] = df['Close/Last'].astype(float)
```

```python
df.info()
```

```python
df['High'] = df['High'].apply(lambda s: s.replace('$', ''))
```

```python
df['High'] = df['High'].astype(float)
```

```python
import matplotlib.pyplot as plt
```

```python
df[['Close/Last', 'High']].plot(kind='box', figsize=(10,3))
```
